```python
# Complete Code to look for a specific string in the covid-19 genom. Group 2.

# Imports
from google.colab import drive
import pprint

# Mounting google drive to load file
drive.mount('/content/drive')
# The following path needs to point to the file saved in the owner's google drive
# Change after /MyDrive to your specific location.
with open('/content/drive/MyDrive/2022-2/Algoritmos/Lab 2/Sequence_Covid-19.fasta') as f: 
    lines = f.readlines()
print("File loaded")

# Parsing the genom as a string
secuence=''
for l in lines[1:]:
  secuence+=l[:-1]
pprint.pprint(secuence)
print("Sequence loaded")

# Searching the target string in the genom. Using brute force.
times_found = 0
comparing_index = 0
target = 'AAAATTTATTGTCC'
location = []
for i in range(len(secuence)):
  if target[comparing_index] == secuence[i]:
    comparing_index += 1
  else:
    comparing_index = 0
  if comparing_index == len(target):
    times_found += 1
    comparing_index = 0
    location.append(i - len(target))
print("The strig was found exactly the following times in the covid-19 genom")
print(times_found)
print("Location of the string starting to count from 0")
print(location)
print("Location of the string starting to count from 1")
print("[{}]".format(location[0] + 1))
```
