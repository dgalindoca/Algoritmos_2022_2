```python
# Complete Code to look for a specific string in the covid-19 genom. Group 2.

# Imports
from google.colab import drive
import pprint

# Mounting google drive to load file
drive.mount('/content/drive')
# The following path needs to point to the file saved in the owner's google drive
# Change after /MyDrive to your specific location.
with open('/content/drive/MyDrive/2022-2/Algoritmos/Lab 2/Sequence_Covid-19.fasta') as f: 
    lines = f.readlines()
print("File loaded")

# Parsing the genom as a string
secuence=''
for l in lines[1:]:
  secuence+=l[:-1]
pprint.pprint(secuence)
print("Sequence loaded")
```

```python
# Searching the target string in the genom. Using brute force (Version 1)

times_found = 0 #Target counter var
comparing_index = 0 #Comparison counter var
target = 'AAAATTTATTGTCC' #Target group 2
location = [] # Positions array
#All this block is O(1) because assign a value is from this order

for i in range(len(secuence)): #O(len(secuence)) = O(n)
  if target[comparing_index] == secuence[i]: #Compare i index of the secuence with the comparing_index from the target
    comparing_index += 1
  else:
    comparing_index = 0
  #O(1)
  if comparing_index == len(target): #Condition to append to location
    times_found += 1
    comparing_index = 0
    location.append(i - len(target)+1)
  #O(1)
#The order of this loop is O(n) (O(n)*O(1) = O(n))

print("The strig was found exactly the following times in the covid-19 genom: ", end = ' ')
print(times_found)
print("Location of the string starting to count from 0", end = ' ')
print(location)
for i in range(len(location)):
  location[i] += len(target)-1
print("End location of the string", location)
#O(1) print
#O(len(location)) final loop

#The total cost of the algorithm is O(n)
```

```python
# Searching the target string in the genom. Using brute force (Version 2)

times_found = 0 #Same as last algorithm
target = 'AAAATTTATTGTCC' #Target group 2
location = [] #Same as last algorithm
#All this block is O(1) because assign a value is from this order

for i in range(len(secuence)): #O(len(sequence)) = O(n)
  if secuence[i:i+len(target)] == target: #O(1)
    times_found += 1
    location.append(i)
#The order of this loop is O(n)*O(1) = O(n)

print("The strig was found exactly the following times in the covid-19 genom: ", end = ' ')
print(times_found)
print("Locations of the string", end = ' ')
print(location)
for i in range(len(location)):
  location[i] += len(target)-1
print("End locations of the string", location)
#O(1) print
#O(len(location)) final loop

#The total cost of the algorithm is O(n)
```
